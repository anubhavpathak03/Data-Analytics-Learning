```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hacktrain = pd.read_csv('hacktrain.csv')
```

```python
hacktrain.head()
```

```python
sns.heatmap(hacktrain.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

```python
hacktrain.isnull().sum()
```

```python
hacktrain.fillna(hacktrain.mean(numeric_only = True), inplace = True)
```

```python
hacktrain.isnull().sum()
```

```python
sns.heatmap(hacktrain.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

```python
hacktrain.head()
```

```python
from scipy.signal import savgol_filter

# Step 1: NDVI columns
ndvi_cols = [col for col in hacktrain.columns if '_N' in col]

# Step 2: Apply smoothing row-wise and re-wrap as DataFrame
smoothed_ndvi = hacktrain[ndvi_cols].apply(
    lambda row: savgol_filter(row, window_length=5, polyorder=2),
    axis=1, result_type='expand'  # ensure output stays aligned as DataFrame
)

# Step 3: Assign back
smoothed_ndvi.columns = ndvi_cols  # keep column names same
hacktrain[ndvi_cols] = smoothed_ndvi

```

```python
hacktrain.head()
```

```python
# Step 1: Feature DataFrame banate hain
features = pd.DataFrame()

# Step 2: Feature extraction from NDVI time-series
features['ndvi_mean'] = hacktrain[ndvi_cols].mean(axis=1)
features['ndvi_std'] = hacktrain[ndvi_cols].std(axis=1)
features['ndvi_min'] = hacktrain[ndvi_cols].min(axis=1)
features['ndvi_max'] = hacktrain[ndvi_cols].max(axis=1)
features['ndvi_range'] = features['ndvi_max'] - features['ndvi_min']

# Step 3: Trend feature (slope of linear regression line)
x = np.arange(len(ndvi_cols))  # time index: 0,1,...,26
features['ndvi_trend'] = hacktrain[ndvi_cols].apply(lambda row: np.polyfit(x, row, 1)[0], axis=1)
```

```python
features.head()
```

```python
hacktrain.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Step 1: Labels (target) ko alag karo
labels = hacktrain['class']   # y

# Step 2: Train-test split
X_train, X_val, y_train, y_val = train_test_split(features, labels, test_size=0.2, random_state=42)

# Step 3: Parameter grid define karo
param_grid = {
    'C': [0.0001, 0.001, 0.009, 0.01, 0.1, 1, 10,20,30,40,50],                 # Regularization strength
    'solver': ['lbfgs', 'saga'],            # Solvers
    'multi_class': ['multinomial']          # Required for multiclass
}

# Step 4: Model & GridSearchCV setup
logreg = LogisticRegression(max_iter=1000)
grid_search = GridSearchCV(logreg, param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Step 5: Fit grid search (train + tune)
grid_search.fit(X_train, y_train)

# Step 6: Best model aur uska score print karo
print("Best Parameters:", grid_search.best_params_)
print("Best CV Accuracy:", grid_search.best_score_)

# Step 7: Final model with best parameters
best_model = grid_search.best_estimator_

# Step 8: Predict on validation set
y_pred = best_model.predict(X_val)

# Step 9: Accuracy check
print("Validation Accuracy:", accuracy_score(y_val, y_pred))
print("Classification Report:\n", classification_report(y_val, y_pred))

```

```python
# Step 1: Load test data
hacktest = pd.read_csv("hacktest.csv")

# Step 2: NDVI columns list banao
ndvi_cols = [col for col in hacktest.columns if '_N' in col]

# Step 3: Feature engineering (same as training)
test_features = pd.DataFrame()
test_features['ndvi_mean'] = hacktest[ndvi_cols].mean(axis=1)
test_features['ndvi_std'] = hacktest[ndvi_cols].std(axis=1)
test_features['ndvi_min'] = hacktest[ndvi_cols].min(axis=1)
test_features['ndvi_max'] = hacktest[ndvi_cols].max(axis=1)
test_features['ndvi_range'] = test_features['ndvi_max'] - test_features['ndvi_min']

x = np.arange(len(ndvi_cols))  # time index: 0,1,...,26
test_features['ndvi_trend'] = hacktest[ndvi_cols].apply(lambda row: np.polyfit(x, row, 1)[0], axis=1)
```

```python
# Step 4: Predict
test_preds = best_model.predict(test_features)

# Step 5: Submission
submission = pd.DataFrame({
    'ID': hacktest['ID'],
    'class': test_preds
})

submission.to_csv("submission.csv", index=False)
```
