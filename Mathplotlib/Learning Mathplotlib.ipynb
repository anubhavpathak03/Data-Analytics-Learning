```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x= [0,1,2,3,4]
y= [0,2,4,6,8]

# Resize the Graph
plt.figure(figsize = (5,3), dpi = 300) 
# dpi no .of pixel here figure size is (5,3) so no of pixels (5x300, 3x300) = (1500, 900)

# plt.plot(x,y, label = '2x', color = 'red', linewidth = 2, marker = '.', markersize = 12, linestyle = '--', markeredgecolor = 'blue')

# shorthand notation 
# fmt = '[color][marker][line]'
plt.plot(x, y, 'r^--', label = '2x', markersize = 12, markeredgecolor = 'blue')

# line number 2
x2 = np.arange(0, 4.5, 0.5)
print(x2)
print(x2[:4])
plt.plot(x2, x2**2, 'g.--', label = 'x^2', markersize = 10)
plt.plot(x2[:4], x2[:4]**2, 'b.--', label = 'x2[:4]', markersize = 10)


plt.title('Our First Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis ---->')
plt.ylabel('Y Axis ---->')

plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8,10])

#  save the graph 
plt.savefig('mygraph.png', dpi = 300)


plt.legend()
plt.show()
```

# Vectorization & Without Vectorization

```python
import time
```

```python
w = np.array([1.0,2.0,3.0,4.0,6.0,7.0,18.0,19.0,17.8,20.9,70.8,60.5])
x = np.array([5.0,7.0,9.0,8.0,10.0,47.0,38.0,49.0,87.8,90.9,117.8,120.5])
b = 4
```

```python
# vectorization 
start_time = time.time()

f = np.dot(w,x) + b

end_time = time.time()

print("Result: ", f)
print("Time taken (in seconds):", end_time - start_time)
```

```python

start_time = time.time()

# without vectorization
f = 0
for j in range(0, 12):
    f = f + (w[j] * x[j])

f = f + b

end_time = time.time()

print("Result: ", f)
print("Time taken (in seconds):", end_time - start_time)
```

```python
## above array size is very small now check 
arr1 = np.random.rand(10_000_000)
arr2 = np.random.rand(10_000_000)
b = 4


start_time = time.time()
# vectorization 
f = np.dot(arr1, arr2) + b

end_time = time.time()

print("by Vectorization: ", f)
print("Time taken (in second during vectorization): ", end_time - start_time)


start_loop = time.time()

# without vectorization
f = 0
for j in range(0, 12):
    f = f + (w[j] * x[j])

f = f + b

end_loop = time.time()

print("by without Vectorization: ", f)
print("Time taken (in second during without vectorization): ", end_loop - start_loop)
```
